```python
JS_output = open("/Users/tarunruchandani/Desktop/HarvardSummer2016/Infomap/JS_output3.txt", 'r')
jsm_list = list(JS_output)
JS_output.close()

jsm_list1=jsm_list[0]
ebunch=jsm_list1
jsm_list = ast.literal_eval(ebunch)

# Build graph out of the JS Similarity Matrix

JS_Sim_Graph = nx.Graph()
JS_Sim_Graph.add_weighted_edges_from(jsm_list)
nx.draw(JS_Sim_Graph,pos=nx.spring_layout(JS_Sim_Graph,iterations=100))
plt.draw()
```
